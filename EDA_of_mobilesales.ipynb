
Exploratory Data Analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Distribution of 'Sale Price'
plt.figure(figsize=(8, 6))
sns.histplot(data['Price'], kde=True,bins=10)
plt.title('Distribution of Sale Price')
plt.xlabel('Sale Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Relationship between 'Sale Price' and 'Quantity'
plt.figure(figsize=(8, 6))
sns.barplot(x='TYPE OF PRODUCT', y='Quantity', data=data,palette="viridis")
plt.title('Sale Price vs. Quantity')
plt.xlabel('Sale Price')
plt.ylabel('Quantity')
plt.show()
```

```python

plt.figure(figsize=(12, 6))
sns.histplot(x='TYPE OF PRODUCT', data=data)
plt.title('Count of Each Product')
plt.xlabel('Product')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
#pie of payment
plt.figure(figsize=(8, 8))
payment_counts = data['PAYMENT TYPE'].value_counts()
plt.pie(payment_counts, labels=payment_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Payment Method Distribution')
plt.show()
```

```python

payment_counts = data['TYPE OF PRODUCT'].value_counts()
plt.pie(payment_counts, labels=payment_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Payment Method Distribution')
plt.show()
```

```python

plt.figure(figsize=(8, 6))
sns.barplot(x='F.Y', y='Quantity', data=data)
plt.title('Year vs. Quantity')
plt.xlabel('Year')
plt.ylabel('Quantity')
plt.show()
```
